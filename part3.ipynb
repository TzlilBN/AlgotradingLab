```python
import pandas as pd
import os

qqq_1 = pd.read_csv('C:\Algo_lab\Final_task\QQQ2023\qqq_eod_202301.txt')
directory = 'C:\Algo_lab\Final_task\QQQ2023'

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):  # Check if the file is a CSV file
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)

# Concatenate all DataFrames into one
options_df = pd.concat(dfs, ignore_index=True)

```

```python
options_df.columns
```

```python
options_df.columns = options_df.columns.str.strip('[] ')

for col in options_df.columns:
    if col.startswith('C_') or col.startswith('P_'):
        options_df[col] = pd.to_numeric(options_df[col], errors='coerce')

dtypes = {
    'QUOTE_UNIXTIME': 'int64',
    'QUOTE_READTIME': 'datetime64[ns]',
    'QUOTE_DATE': 'datetime64[ns]',
    'QUOTE_TIME_HOURS': 'float64',
    'UNDERLYING_LAST': 'float64',
    'EXPIRE_DATE': 'datetime64[ns]',
    'EXPIRE_UNIX': 'int64',
    'DTE': 'float64',
    'C_DELTA': 'float64',
    'C_GAMMA': 'float64',
    'C_VEGA': 'float64',
    'C_THETA': 'float64',
    'C_RHO': 'float64',
    'C_IV': 'float64',
    'C_VOLUME': 'float64',
    'C_LAST': 'float64',
    'C_SIZE': 'float64',
    'C_BID': 'float64',
    'C_ASK': 'float64',
    'STRIKE': 'float64',
    'P_BID': 'float64',
    'P_ASK': 'float64',
    'P_SIZE': 'float64',
    'P_LAST': 'float64',
    'P_DELTA': 'float64',
    'P_VOLUME': 'float64',
    'STRIKE_DISTANCE': 'float64',
    'STRIKE_DISTANCE_PCT': 'float64'
}

options_df = options_df.astype(dtypes)
```

```python
# Remove rows with low  or Nan volume
options_df.drop(options_df[(options_df['C_VOLUME'] < 100) | options_df['C_VOLUME'].isna()].index, inplace=True)
options_df.drop(options_df[(options_df['P_VOLUME'] < 100) | options_df['P_VOLUME'].isna()].index, inplace=True)

```

```python
# Add calculation columns

options_df['PCP'] = options_df['UNDERLYING_LAST'] - options_df['STRIKE'] + options_df['P_LAST']
options_df['PCP_PROFIT'] = options_df['C_LAST'] - options_df['PCP']
options_df
```

```python
options_df['ARB_PROFIT'] = 0

options_df.loc[options_df['PCP_PROFIT'] > 0, 'ARB_PROFIT'] = (options_df['C_BID'] - (options_df['UNDERLYING_LAST'] - options_df['STRIKE'] + options_df['P_ASK'])) * 100
options_df.loc[options_df['PCP_PROFIT'] < 0, 'ARB_PROFIT'] = ((options_df['UNDERLYING_LAST'] - options_df['STRIKE'] + options_df['P_BID']) - options_df['C_ASK']) * 100

options_df
```

```python

options_df['QUOTE_DATE'] = pd.to_datetime(options_df['QUOTE_DATE'])
# Find max arb profit per day 
max_profit_per_day = options_df.groupby('QUOTE_DATE')['ARB_PROFIT'].max().reset_index()
max_profit_per_day = max_profit_per_day.merge(options_df, on=['QUOTE_DATE', 'ARB_PROFIT'], how='left')
# Extract desired columns
desired_columns = ['QUOTE_DATE', 'C_LAST', 'PCP', 'PCP_PROFIT', 'ARB_PROFIT']
max_profit_per_day = max_profit_per_day[desired_columns].copy()
max_profit_per_day
```

```python
max_profit_per_day = max_profit_per_day.drop_duplicates(subset=['QUOTE_DATE'])
max_profit_per_day
```
